## Exercise 1 

Two dices are thrown once and the total score is observed. Use a simulation to find the estimated probability that the total score is even or greater than 7.  A simulation is a repetition of the same experiment multiple times to observe its behavior:

- Run the experiment 1000 times (roll 2 dice 1000 times, and sum the number of both dices).
- Keep track of the number of times that the sum was either greater than 7 or an even number.
- Divide the number from step 2 by the number of iterations (1000).

```python
import numpy as np
import pandas as pd

# Roll dice once
def roll_dice():
    return np.random.randint(1,7)
# Sum number of dices
def sum_dices(n):
    score = 0
    for i  in range(n):
        score = score + roll_dice()
    return score
# Running the experiment n times
# Track of greater than 7
# Track of evens
def experiment(times):
    track = []
    greater_seven = 0
    even = 0
    both_or = 0
    for i in range(times): 
        track.append(sum_dices(2))
        if track[i] > 7:    # greater than seven
            greater_seven += 1
            #print("Greater",greater_seven)
        if (track[i] % 2) == 0: #even
            even += 1
            #print("Even",even)
        if (track[i] > 7) or ((track[i] % 2) == 0):
            both_or += 1

    prob_greater_seven = (greater_seven/times)*100
    prob_even = (even/times)*100
    prob_both = (both_or/times)*100

    return track, prob_greater_seven , prob_even, prob_both

track, prob_greater_seven, prob_even, prob_both = experiment(1000)
#Print
print(f'Probability greater than 7: {prob_greater_seven:.2f}%')
print(f'Probability even: {prob_even:.2f}%')
print(f'Probability of greater than 7 or probability even:{prob_both:.2f}%')
print(f'Track of values: {track}')
```

## Exercise 2

A box contains 10 white balls, 20 red balls and 30 green balls. If we take 5 balls from the box with replacement (we take the ball, observe what color it is and put it back into the box). We want to know the probability of:

1. Take 3 white and 2 red.
2. Take all of the same color.

Run the experiment 1000 times and calculate the above probabilities.

```python
ball_box = {}

# Create the box of balls
for i in range(60):
    if i < 10:
        ball_box[i] = "White"
    elif (i > 9) and (i < 30):
        ball_box[i] = "Red"
    else:
        ball_box[i] = "Green"

#print(ball_box)
            
# TODO
# Take 1 ball
def take_ball():
    i = np.random.randint(0,60)
    return i, ball_box[i]

# Take n balls
def take_ball_n_times(n):
    track_color = []
    for _ in range(n):
        track_index, color_ball = take_ball()
        track_color.append(color_ball)
    return track_color
# Experiment n times
def experiment(n):  # 3 White & 2 Red
    # Generate an array from the list take_ball_n_times
    # Generate an array from that array. 2D
    return np.array([np.array(take_ball_n_times(5)) for _ in range(n)])   
    # track1 = []
    # for i in range(n):
    #     track1.append(take_ball_n_times(5))
    #     print(track1[i][1])


# Probability 3 White & 2 Red
# Real probability
# Prob (3 White & 2 Red)
# prob(w&w&w&r&r) + prob(r&w&w&w&r) + prob(r&r&w&w&w) + ...
# p(w)*p(w)*p(w)*p(r)*(r) + p(r)*p(w)*p(w)*p(w)*(r) + ... 
experiment = experiment(1000)
def prob_three_white_two_red(n, experiment):
    # Iterate over experiment results
    aux = 0
    for i in range(n):  # rows
        red = 0
        white = 0
        for j in range(5): # columns
            if(experiment[i][j] == 'Red'):
                red += 1
            elif(experiment[i][j] == 'White'):
                white += 1
        if(white == 3 & red == 2):
            aux += 1 
    #End of experiment
    prob_experiment1 = (aux/n)*100 
    return prob_experiment1

# Probability of all being the same color
def prob_all_same_color(n, experiment):
    n_whites = 0
    n_reds = 0
    n_greens = 0
    for i in range(n):  # rows
        white = 0
        red = 0
        green = 0
        for j in range(5): # columns
            if(experiment[i][j] == 'White'):
                white += 1
            elif(experiment[i][j] == 'Red'):
                red += 1
            elif(experiment[i][j] == 'Green'):
                green += 1
        # When the row is complete, we check the colors
        if(white == 5):
            n_whites += 1 
        elif(red == 5):
            n_reds += 1 
        elif(green == 5):
            n_greens += 1 
    # Probability of each color    
    prob_whites = (n_whites/n)*100 
    prob_reds = (n_reds/n)*100
    prob_greens = (n_greens/n)*100
    
    return prob_whites + prob_reds + prob_greens

# Print
print(f'Probability 3 White & 2 Red: {prob_three_white_two_red(1000,experiment = experiment):.2f}%')
print(f'Probability of all being the same color: {prob_all_same_color(1000, experiment = experiment):.2f}%')
```
